```python
import pandas as pd
import folium
from IPython.display import display

# Function to create a map with latitude, longitude, and an additional variable from a CSV file
def create_map_from_csv(ProviderData, lat, long, Industries):
    # Load the CSV file into a DataFrame with the specified encoding
    df = pd.read_csv("ProviderData.csv", encoding='latin1')
    
    # Create a map centered at the average latitude and longitude
    m = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=6)
    
    # Define colors for each industry
    industry_colors = {
        'Automotive & Transportation': 'blue',
        'Healthcare': 'red',
        'Information Technology': 'green',
        'Construction': 'purple',
        'Manufacturing & Logistics': 'orange',
        'Business Financial Services': 'darkred',
        'Transportation': 'darkblue',
        'Other': 'black'
    }
    
    # Add markers for each row in the DataFrame
    for _, row in df.iterrows():
        color = industry_colors.get(row['Industries'], 'gray')  # Default to gray if industry not found
        folium.Marker(
            location=[row['lat'], row['long']],
            popup=f"{'Industries'}: {row['Industries']}",
            icon=folium.Icon(color=color)
        ).add_to(m)
    
    # Add legend to the map
    legend_html = '''
     <div style="position: fixed; 
     bottom: 50px; left: 50px; width: 200px; height: 300px; 
     border:2px solid grey; z-index:9999; font-size:14px;
     background-color:white;
     ">
     <b>Industries Legend</b><br>
     <i class="fa fa-map-marker fa-2x" style="color:blue"></i>Automotive & Transportation<br>
     <i class="fa fa-map-marker fa-2x" style="color:red"></i>Healthcare<br>
     <i class="fa fa-map-marker fa-2x" style="color:green"></i>Information Technology<br>
     <i class="fa fa-map-marker fa-2x" style="color:purple"></i>Construction<br>
     <i class="fa fa-map-marker fa-2x" style="color:orange"></i>Manufacturing & Logistics<br>
     <i class="fa fa-map-marker fa-2x" style="color:darkred"></i>Business Financial Services<br>
     <i class="fa fa-map-marker fa-2x" style="color:darkblue"></i>Transportation<br>
     <i class="fa fa-map-marker fa-2x" style="color:black"></i>Other<br>
      </div>
     '''
    m.get_root().html.add_child(folium.Element(legend_html))
    
    # Display the map
    display(m)

# Example usage
csv_file = 'Provider_Data_1.csv' 
lat_col = 'lat'  
lon_col = 'long'  
additional_col = 'Industries'  

create_map_from_csv("ProviderData.csv", 'lat', 'long', 'Industries')
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('ProviderData.csv', encoding='latin1')

# Set the style for seaborn
sns.set(style="whitegrid")

# Create an area plot for Cost vs Frequency for different Industries in one chart
plt.figure(figsize=(12, 8))

industries = df['Industries'].unique()
for industry in industries:
    subset = df[df['Industries'] == industry]
    sns.kdeplot(subset['Cost'], shade=True, label=industry)

plt.xlabel('Cost')
plt.ylabel('Frequency')
plt.title('Area Plot of Cost vs Frequency for Different Industries')
plt.legend(title='Industries')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('ProviderData.csv',  encoding='latin1')

# Count the number of providers in each industry
industry_count = df['Industries'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = industry_count.plot(kind='bar')
plt.xlabel('Industry')
plt.ylabel('Frequency')
plt.title('Number of Providers in Each Industry')
plt.xticks(rotation=45)

# Add the count on top of each bar
for bar in bars.patches:
    plt.annotate(format(bar.get_height(), '.0f'),
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='center',
                 size=10, xytext=(0, 8),
                 textcoords='offset points')

plt.show()
```
